# Simple Linear Regression - Supervised Model

```python
###### Simple Linear Regression - Supervised Model
# Parametrik algoritmadır. (Matematiksel bir denklem ile ifade edilen)
# Continuous değerler tahmin edilmeye çalışılır
# Simple ve Multi olarak ikiye ayrılır
# Y=b0+b1X (basic regression equation)
# b0: intercept        : X0 ın  y eksenini kestiği nokta
# b1: slope            : Eğim
# ei = yi - y(head)i   : Hata
# y(head) = b0 + b1*X  : Regression Equation
# Örnek: Salary = b0 + b1*Experience


# simple linear regression: Y=bo+b1X
# multiple linear regression: y=bo+b1*x1+b2*x2+.....+bn*xn
```

## Linear Regression

```python
sns.regplot(x="total_spend", y="sales", data=df, ci=None)  # Eldeki data icin best fit i cizer
```

## Splitting the dataset into X(independent variables) and y (dependent variable)

```python
# y_pred = b1X + b0
```

```python
X= df["total_spend"]
y= df["sales"]
```

## Determination of coefficients (slope and intercept)

```python
np.polyfit(X, y, deg=1) #Bulunan ilk değer b1 (slope), ikinci değer b0 (intercept)' dır.
```

```python
slope, intercept = np.polyfit(X, y, deg=1)
```

```python
print("slope    :", slope)
print("intercept:", intercept)  
```

```python
#y=ax+b  
#y=0.04X+4.24  ----->bu formül ile X leri verdiğimiz zaman y'yi tahmin edebilmiş olacaz 
#bu simple linear regression için geçerli multiple linear regression da farklı olacak
#y=bo+b1*x1+b2*x2+.....+bn*xn
```

## Why do we use the least squares error method to find the regression line that best fits the data?

```python
#Her bir gözlem için hatayı hesaplayıp minimize etmeye çalışıyoruz
#Hataları minimize ederek en iyi doğruyu(best fit line ı)çizmiş oluyoruz
```

```python
b1, b0 = np.polyfit(X, y, deg=1)
```

```python
y_pred = b0 + b1 * X
```

```python
values = {'actual' : y, 'predicted' : y_pred, 'residual' : (y - y_pred), 'LSE' : (y - y_pred) ** 2}
df_2 = pd.DataFrame(values)
df_2
#residual:gerçek değer ile(y) tahmin ettiğimin değer arasındaki farktır(y_pred)
```

```python
df_2.residual.sum().round() #Datamız Linear Regression a uygunsa residual toplamı sıfır olacak
```
